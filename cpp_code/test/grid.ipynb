```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N_particles = 12

L=np.ceil( np.sqrt(N_particles))

# Create a grid of points
x=np.zeros(N_particles)
y=np.linspace(0,N_particles,N_particles)

for i in range(N_particles):
	x[i]=i%L
	y[i]=np.floor(i/L)

plt.scatter(x,y)


```
